```julia
using JLD2
using Plots

plotg = JLD2.load("lossg.jld2")["lossg_vals"]
first = findfirst(isequal(0), plotg)
if isnothing(first)
    first = length(plotg)
end
plot(plotg[1:length(plotg)], title = "Generator Loss", linewidth=2, legend = false)
```

```julia
plotd = JLD2.load("lossd.jld2")["lossd_vals"]
first = findfirst(isequal(0), plotd)
if isnothing(first)
    first = length(plotd)
end
plot(plotd[1:first], title = "Discriminator Loss", linewidth=2, legend = false)
```

```julia
plotnorm = JLD2.load("gradnorms.jld2")["grad_norms"]
first = findfirst(isequal(0), plotnorm)
if isnothing(first)
    first = length(plotnorm)
end
plot(plotnorm[1:first], title = "Norm of Gradients of PQC", linewidth=2, legend = false)
```
